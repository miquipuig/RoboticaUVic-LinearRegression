```python
import numpy as np
data = np.array([map(float, l.split()) for l in open('data.txt')])
y = data[:,-1]
print 'Data points', len(y)
print y.T
```

```python
# First define some handy shortcuts 
dot = np.dot
inv = np.linalg.inv
# Dummy features
X = np.ones((28,1))
print X
# Fitting the parameters: theta = (X'*X)^-1*X'*y
theta = dot(dot(inv(dot(X.T, X)), X.T), y)
print theta
```

```python
# MSE = (1/N)*sum((y-X*theta)^2) 
print sum((y-dot(X, theta))**2) / len(y)
```

```python
# Let's add a continuous variable, like the area of the house
print X
X = np.hstack((X, X))
print X
print X.shape
```

```python
from scipy.linalg import lstsq
theta = lstsq(X,y)[0]
print sum((y-dot(X, theta))**2) / len(y)
```

```python
# Let's add the number of garages as anoher variable
X = np.hstack((X, data[:,4].reshape(len(y),1)))
theta = lstsq(X,y)[0]
print sum((y-dot(X, theta))**2) / len(y)
```

```python
#To add a categorical variable, like "construction type" we have to decide how to represent it. Two options: 
# 1) Giving a value to every category choice (1, 2, 3, ...).
# 2) Using an indicator vector, with a zero for each dimension, except for the active that is one.
X1 = np.hstack((X, data[:,8].reshape(len(y),1)))
ind = np.zeros((len(y), max(data[:,8]).astype(int)))
ind[range(0,len(y)), (data[:,8]-1).astype(int)]=1
X2 = np.hstack((X, ind))
print 'Representation 1,2,3,4,...', '\n', X1
print 'Representation [0,1,0,0]', '\n', X2
```

```python
theta = lstsq(X1,y)[0]
print 'Representation 1,2,3,...', sum((y-dot(X1, theta))**2) / len(y)
theta = lstsq(X2,y)[0]
print 'Representation [0,1,0,0]', sum((y-dot(X2, theta))**2) / len(y)
# Just one word of caution with indicator vectors: if we use more dimensions 
# than we should, we may end up with a low-rank matrix that we can't invert!
```

```python
import numpy
data = numpy.array([map(float, l.split()) for l in open('../housing.data')])
data=numpy.loadtxt('../housing.data')

#separem última columna
y = data[:,-1]
mitjaY= numpy.mean(y)
var=sum((mitjaY-y)**2) / len(y)
print "Mitja:", mitjaY
print "Variança general: ", var
MSE= sum((y-mitjaY)**2) / len(y)
print "MSE amb la mitja com a model: ", MSE
#Partim les dades
trainingData = data[0:len(y)/2,:]
yTraining= trainingData[:,-1]
testingData =data[len(y)/2:,:]
yTesting= testingData[:,-1]
dot = numpy.dot
inv = numpy.linalg.inv

alf= 0.00000001
for i in range(0, len(trainingData[1,:])-1):
    X=trainingData[:,i:i+1]
    theta = dot(dot(inv(dot(X.T, X)), X.T), yTraining)
    MSEtraining= sum((yTraining-dot(X, theta))**2) / len(yTraining) 
    X=testingData[:,i:i+1]
    MSEtesting= sum((yTesting-dot(X, theta))**2) / len(yTesting)
    R2=1-(MSEtesting/var)
    print "----------------------------------------------------------------"
    print "theta per la columna ",i, ": ", theta
    print "MSE training: ",MSEtraining
    print "MSE testing: ", MSEtesting
    print "R^2 de testing: ", R2

print "----------------------------------------------------------------"
print "La millor columna la 5 (comptant que la primera es la 0) i la pitjor és la 0 doncs te el major error. "
print "Es la que ens genera menys error per si sola i per tant per separa mes informativa "   
    
    
#Incluir totes les columnes
X=trainingData
theta = dot(dot(inv(dot(X.T, X)), X.T), yTraining)
MSEtraining= sum((yTraining-dot(X, theta))**2) / len(yTraining)+ alf*(abs(sum(theta**2)))**(1/2)
X=testingData
MSEtesting= sum((yTesting-dot(X, theta))**2) / len(yTesting)+alf*(abs(sum(theta**2)))**(1/2)
R2=1-(MSEtesting/var)

```

```python
print "# Q4:"
print "Theta incloent totes les files: ", theta
print "MSE training: ",MSEtraining
print "MSE testing: ", MSEtesting
print "R^2 de testing: ", R2
print "Obtenim un predictor perfecte"

X=trainingData[:,1:-1]
theta = dot(dot(inv(dot(X.T, X)), X.T), yTraining)
MSEtraining= sum((yTraining-dot(X, theta))**2) / len(yTraining)+ alf*(abs(sum(theta**2)))**(1/2)
X=testingData[:,1:-1]
MSEtesting= sum((yTesting-dot(X, theta))**2) / len(yTesting)+alf*(abs(sum(theta**2)))**(1/2)
R2=1-(MSEtesting/var)
print " "
print "Theta obviant la primera: ", theta
print "MSE training: ",MSEtraining
print "MSE testing: ", MSEtesting
print "R^2 de testing: ", R2
print "El predictor empitjora moltíssim"
```
